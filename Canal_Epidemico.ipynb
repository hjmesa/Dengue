```python

# Cargar el archivo de datos historicos

FILE = 'local/data/datos_mensuales_consolidados_cali_nino_2010_2016.csv'

dataset = pd.read_csv(FILE, header=0, index_col=False)


#Eliminar año y radiacion solar minima que siempre es cero (noche)
dataset = dataset.drop(["Sin_ajustar", "Sospechoso", "Probable", "Confirmado_Lab", "Confirmado_Cli", "Confirmado_Nexo", "Todos"], axis=1)


# Se carga un arreglo con todos los datos

values = dataset.values
values = values.astype('float64')

Canal_T = np.zeros((12,10))
Datos = np.zeros((9))
k=0
for i in range (0,10):
    for j in range (0,12):
        Canal_T[j][i]=values[k][14]
        k=k+1

j=0
for i in range (k,k+9):
    Datos[j]=values[i][14]
    j=j+1
    

print (Canal_T)
Canal = Canal_T.transpose()
print (Canal)
print (Datos)
dataset
```
